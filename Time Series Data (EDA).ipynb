# Question 1 and 2

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime as datetime
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
df = pd.read_csv("KCLT.csv")
df['date']= pd.to_datetime(df['date'])
```

# Question 3

```python
def moving_avg(column, period):
    weight = np.random.random(period)
    weight /= weight.sum() 
    moving_avg = np.convolve(column ,weight,'valid')
    return moving_avg

col = df.actual_mean_temp
wma = moving_avg(col,5)
print(wma)
```

# Question 4.1(Q5)

```python
plt.figure(figsize = (16,5))
plt.legend((df["actual_mean_temp"],df["record_min_temp"]))
plt.plot(df['date'], df["actual_mean_temp"], label='Actual Average Temperature', color = 'red')
plt.plot(df['date'], df["record_min_temp"], label='Record Lowest Temperature', color = 'purple')
plt.legend()
plt.title('Temperature during the Year')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.show()
```

# Question 4.2(Q5)

```python
plt.figure(figsize = (15,5))
plt.legend((df["record_max_temp"],df["actual_max_temp"]))
plt.plot(df["record_max_temp"], label='Record Maximum Temperature', color = 'forestgreen')
plt.plot(wma, label = "Weighted Average Temprature", color = 'purple')
plt.plot(df["actual_max_temp"], label='Actual Maximum Temperature', color = 'red')
plt.legend()
plt.title('Temperature during the year')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.show()
```

In the first graph we can observe actual average temerature in December 2014 is higher than other winter months although the lowest record temerature is recorded in the same month of the year.
